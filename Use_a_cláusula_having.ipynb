

# **1.   cláusula having**



Quando a condição é acompanhada de uma função (sum, avg, min, max, count, ...) utilizamos o comando HAVING em vez do WHERE.


    SELECT column_name(s) FROM table_name GROUP BY column_name(s) HAVING condition

Exemplo:


    SELECT COUNT(id_cliente), id_loja FROM transacoes GROUP BY id_loja HAVING COUNT(id_cliente) > 1;
